Create the class AntBaitCSV:
* define read all csv file function
* define read part csv file function
* define read each row of csv file and store to list function
* define read each attribute and store to list function 

```python
class AntBaitCSV(object):

    def __init__(self):
        self.month = []
        self.year = []
        self.plot = []
        self.stake = []
        self.species = []
        self.abundance = []

    def ReadFullCSV(self, filename):
        FullCsv = pd.read_csv(filename, usecols=['species', 'month', 'year', 'abundance', 'plot', 'stake'])
        print(FullCsv)

    def ReadPartCSV(self, filename):
        PartCsv = pd.read_csv(filename, usecols = ['species', 'month', 'year', 'abundance'])
        print(PartCsv)

    def ReadCSVtoList(self,filename):
        file = open(filename, 'r', encoding="gbk")
        context = file.read()
        list_result = context.split("\n")
        length = len(list_result)
        for i in range(length):
            list_result[i] = list_result[i].split(",")
            print("ant_bait_list", i, "=", list_result[i])

    def ReadBait_species_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            species = [row['species'] for row in reader]
        return species

    def ReadBait_month_list(self, filename):
        with open(filename, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            month = [row['month'] for row in reader]
        return month

    def ReadBait_year_list(self, filename):
        with open(filename, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            year = [row['year'] for row in reader]
        return year

    def ReadBait_plot_list(self, filename):
        with open(filename, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            plot = [row['plot'] for row in reader]
        return plot

    def ReadBait_stake_list(self, filename):
        with open(filename, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            stake = [row['stake'] for row in reader]
        return stake

    def ReadBait_abundance_list(self, filename):
        with open(filename, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            abundance = [row['abundance'] for row in reader]
        return abundance
```

Create the class AntSpeciesCSV:

* define read all csv file function
* define read part csv file function
* define read each row of csv file and store to list function
* define read each attribute and store to list function

```python
class AntSpeciesCSV(object):

    def __init__(self):
        self.speciescode = []
        self.genus = []
        self.altgenus = []
        self.species = []
        self.altspecies = []
        self.tribe = []
        self.subfamily =[]
        self.IDissues = []

    def ReadFullCSV(self, filename):
        FullCsv = pd.read_csv(filename)
        print(FullCsv)

    def ReadPartCSV(self, filename):
        PartCsv = pd.read_csv(filename, usecols = ['species', 'genus'])
        print(PartCsv)

    def ReadCSVtoList(self,filename):
        file = open(filename, 'r', encoding= "gbk")
        context = file.read()
        list_result = context.split("\n")
        length = len(list_result)
        for i in range(length):
            list_result[i] = list_result[i].split(",")
            print("ant_species_list", i, "=", list_result[i])

    def ReadSpecies_speciescode_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            speciescode = [row['speciescode'] for row in reader]
        return speciescode

    def ReadSpecies_genus_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            genus = [row['genus'] for row in reader]
        return genus

    def ReadSpecies_altgenus_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            altgenus = [row['altgenus'] for row in reader]
        return altgenus

    def ReadSpecies_species_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            species = [row['species'] for row in reader]
        return species

    def ReadSpecies_altspecies_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            altspecies = [row['altspecies'] for row in reader]
        return altspecies

    def ReadSpecies_tribe_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            tribe = [row['tribe'] for row in reader]
        return tribe

    def ReadSpecies_subfamily_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            subfamily = [row['subfamily'] for row in reader]
        return subfamily

    def ReadSpecies_IDissues_list(self,filename):
        with open(filename,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            IDissues = [row['IDissues'] for row in reader]
        return IDissues
```

Store each attribute in the two csv file to each list
